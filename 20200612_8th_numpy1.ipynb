### numpyでのIndexing,Slicing
行列の中から欲しいデータだけを取りたい時の操作。
例）全ての列の先頭、後ろから10個など

```python
ndarray = np.array([[1,2],[3,4],[5,6],[7,8]])
print(ndarray.shape)
print("")
print(ndarray)
```

```python
"""
indexing:要素の取得
"""
print(ndarray[0][0])
print(ndarray[3][1])
print(ndarray[-1][-1])
print(ndarray[-1,0])
```

```python
"""
Slicing 複数要素の取り出し
[y方向の始点：終点、x方向の始点：終点]
始点と終点のどちらかだけを指定することも可能。
"""

print(ndarray[0:4,0:1])
print("")
print(ndarray[1:,])
print("")
print(ndarray[0:4:2,])
```

### その他：numpyによる画像の切り出し
画像を扱うライブラリとnumpyで、ほとんどの画像処理は実行可能

```python
"""
画像の配列は(height, witdh, channel)の形
白黒の場合は(height, witdh)
"""
#画像処理ライブラリ
import cv2
from PIL import Image
from skimage import io

#パス操作のライブラリ
import os

#グラフ化、画像表示ライブラリ
import matplotlib.pyplot as plt
```

```python
#画像の読み込み
desktop_path = os.path.join(os.path.join(os.environ['USERPROFILE']), 'Desktop')
image_name = "dog.jpg"
img = cv2.imread(os.path.join(desktop_path,image_name))
img_2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
print(img_2.shape)

#画像の表示
plt.imshow(img_2)
```

```python
#画像の切り出し
#範囲を指定するだけで、簡単に切り取れる。
img_cut = img_2[0:200,100:300]

print(img_cut.shape)
plt.imshow(img_cut)
```
