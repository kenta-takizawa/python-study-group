### 問題６：指数関数 y = e^x および y = x^2のグラフを横に並べて表示してください。 
-5 <= x <= 5で表示してください。subplots

```python
#以下のようなグラフを表示するようなプログラムを書いてください。
from IPython.display import Image
Image("./matplotlib_plots/test_6.jpg")
```

```python
#答え
x = np.linspace(-5,5,20)
y = np.exp(x)
y2 = x**2

fig = plt.figure()
ax = fig.add_subplot(121)
ax.plot(x,y,label="no1")

ax2 = fig.add_subplot(122)
ax2.plot(x,y2,label="no2")
plt.savefig("./matplotlib_plots/test_6.jpg",bbox_inches="tight")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[1].plot(x, y2)
```

### 問題７：for文とsubplotを使って、1行4列のグラフを表示してください。

色、関数、ラインスタイルは何でも構いません。凡例はなくてもかまいません。

```python
#以下のようなグラフを表示するようなプログラムを書いてください。
from IPython.display import Image
Image("./matplotlib_plots/test_7.jpg")
```

```python
x = np.linspace(-5,5,20)
y = np.exp(x)
y2 = x**2
y3 = x
y4 = 1/x

colors = ["r","g","b","orange"]
markers = ["o","^","o","."]
linestyles = ["","-","--","-."]
functions = [y,y2,y3,y4]
labels = ["関数" + str(a) for a in range(4)]

fig, axes = plt.subplots(figsize=(10,5),nrows=1, ncols=4)

for i in range(4):
    axes[i].plot(x,functions[i],c = colors[i],
                marker=markers[i],linestyle=linestyles[i],label=labels[i])
    axes[i].set_xlabel("x")
    axes[i].set_ylabel("y"+str(i+1))

fig.legend(bbox_to_anchor=(1, 0.95), loc='upper left', borderaxespad=0)
plt.tight_layout()
plt.savefig("./matplotlib_plots/test_7.jpg",bbox_inches="tight")
```
