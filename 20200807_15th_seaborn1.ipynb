### 問題１：タイタニック号のデータを用いて、ヒストグラムを表示してください。ただし、seabornを用いること。
https://github.com/uehara1414/japanize-matplotlib/issues/1

```python
#以下のようなグラフを表示するようなプログラムを書いてください。
from IPython.display import Image
Image("./seaborn1/test_1.png")
```

```python
import pandas as pd
df = pd.read_csv('./titanic/train.csv')
df = df.dropna(subset=['Age'])

plt.figure()
sns.distplot(df['Age'])
plt.title("乗客の年齢ヒストグラム")
plt.savefig("./seaborn1/test_1.png")
```

### 問題２：年齢のヒストグラムを表示してください。ただし、生存者と死亡者で分けてプロットしてください。

```python
#以下のようなグラフを表示するようなプログラムを書いてください。
from IPython.display import Image
Image("./seaborn1/test_2.png")
```

```python
survive = df[df["Survived"] == 1]
dead = df[df["Survived"] == 0]

plt.figure()
sns.distplot(survive['Age'],label="生存者")
sns.distplot(dead['Age'],label="死亡者")
plt.legend()
plt.title("乗客の年齢ヒストグラム")
plt.savefig("./seaborn1/test_2.png")
```

### 問題３：以下のように2変数間の散布図を表示してください。

```python
#以下のようなグラフを表示するようなプログラムを書いてください。
from IPython.display import Image
Image("./seaborn1/test_3.png")
```

```python
plt.figure()
sns.jointplot(x='Age', y='Fare', data=df,alpha=0.3)
plt.savefig("./seaborn1/test_3.png")
```

### 問題４：以下のように2変数間の散布図を複数表示してください。

```python
#以下のようなグラフを表示するようなプログラムを書いてください。
from IPython.display import Image
Image("./seaborn1/test_4.png")
```

### 問題９：タイタニック号のデータを使って、以下のようにヒートマップを表示してください。
ヒートマップ：変数間の相関を表でまとめたもの。

```python
#以下のようなグラフを表示するようなプログラムを書いてください。
from IPython.display import Image
Image("./seaborn1/test_9.png")
```

```python
import pandas as pd
df = pd.read_csv('./titanic/train.csv')
corr = df.corr()

plt.figure()
sns.heatmap(corr, cmap='coolwarm', annot=True)
plt.tight_layout()
plt.savefig("./seaborn1/test_9.png")
```
