### 問題１：以下のグラフを書いてください。また、グラフの大きさを(20,10)で表示してください。

```python
#以下のようなグラフを表示するようなプログラムを書いてください。
from IPython.display import Image
Image("./matplotlib_plots2/test_1.jpg")
```

```python
#答え
plt.figure(figsize=(20,10))
x = np.linspace(-5,5,20)
y = np.exp(x)

plt.ylabel("y軸",fontsize=18)
plt.xlabel("x軸",fontsize=18)
plt.title("問題１")
plt.plot(x,y,c="g")
plt.savefig("./matplotlib_plots2/test_1.jpg")
```

### 問題２：複数のグラフを作成して、PDFで保存してください。
見本は別紙PDFを確認してください。

```python
##正直、あまり使ったことなかった。
##
# PdfPagesをimport(覚える必要なし，コピペで使い回す)
from matplotlib.backends.backend_pdf import PdfPages
# pdfインスタンス作成．この時点でファイルができる
pdf = PdfPages('./matplotlib_plots2/savefig_sample.pdf')

for i in range(0, 10):
    #------グラフ作成--------
    fig, axes = plt.subplots()
    #グラフの形状が徐々にかわるように設計しました．（適当です．）
    axes.plot(x, y + x*i)
    #タイトルをつけます．pdfで文字検索ができることを確認してください．
    axes.set_title('ID:#{}'.format(i))
    #-----------------------
 
    #forループのなかで保存
    pdf.savefig(fig)
 
#ループ後に閉じます
pdf.close()

```

### 問題３：以下のような散布図を表示してください。
np.sin,np.cosを使います。

```python
#以下のようなグラフを表示するようなプログラムを書いてください。
from IPython.display import Image
Image("./matplotlib_plots2/test_3.jpg")
```

```python
x = np.arange(0,10,0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure()
plt.scatter(sin, cos, alpha=0.6)
plt.title("sin関数とcos関数の散布図")
plt.xlabel("sin関数")
plt.ylabel("cos関数")
plt.savefig("./matplotlib_plots2/test_3.jpg")
```

### 問題５：問題４に続き、年齢ヒストグラムを表示してください。ただし、生き残った乗客と亡くなった乗客を分けて表示してください。

```python
#以下のようなグラフを表示するようなプログラムを書いてください。
from IPython.display import Image
Image("./matplotlib_plots2/test_5.jpg")
```

```python
import pandas as pd
df = pd.read_csv('./titanic/train.csv')

survive = df[df["Survived"]==1]
dead = df[df["Survived"]==0]

plt.figure(figsize=(10,5))
plt.hist(survive['Age'],alpha=0.6,bins=50,label="生存:"+str(len(survive))+"人")
plt.hist(dead['Age'],alpha=0.6,bins=50,label="死亡:"+str(len(dead))+"人")
plt.title("乗客の年齢ヒストグラム")
plt.xlabel("年齢")
plt.legend()
plt.savefig("./matplotlib_plots2/test_5.jpg")
```
