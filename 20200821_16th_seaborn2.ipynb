```python
import seaborn as sns
sns.set(font="IPAexGothic") 
import numpy as np 
import matplotlib.pyplot as plt
import japanize_matplotlib
%matplotlib inline
```

### 問題１：タイタニック号のデータを使って、以下のようにヒートマップを表示してください。
ヒートマップ：変数間の相関を表でまとめたもの。

```python
#以下のようなグラフを表示するようなプログラムを書いてください。
from IPython.display import Image
Image("./seaborn2/test_1.png")
```

```python
import pandas as pd
df = pd.read_csv('./titanic/train.csv')
corr = df.corr()

plt.figure()
sns.heatmap(corr, cmap='coolwarm', annot=True)
plt.tight_layout()
plt.savefig("./seaborn2/test_1.png")
```

### 問題２：以下のように棒グラフを表示してください。

```python
import pandas as pd
df = pd.read_csv('./titanic/train.csv')

plt.figure()
sns.countplot(x='Sex', data=df, hue='Survived')
plt.savefig("./seaborn2/test_2.png")
```

### 問題３：以下のように箱ひげ図グラフを表示してください。

```python
#以下のようなグラフを表示するようなプログラムを書いてください。
from IPython.display import Image
Image("./seaborn2/test_3.png")
```

```python
import pandas as pd
df = pd.read_csv('./titanic/train.csv')

plt.figure()
sns.boxplot(x='Pclass', y='Age', data=df, hue='Survived')
plt.savefig("./seaborn2/test_3.png")
```

### 問題４：以下のように分布グラフを表示してください。

```python
#以下のようなグラフを表示するようなプログラムを書いてください。
from IPython.display import Image
Image("./seaborn2/test_4.png")
```

```python
plt.figure()
sns.swarmplot(x='Pclass', y='Age', data=df, size=4, hue='Survived')
plt.savefig("./seaborn2/test_4.png")
```
