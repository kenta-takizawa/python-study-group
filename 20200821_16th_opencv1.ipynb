### 問題１：次の画像を読み込んで、notebook上に表示してください。また、読み込んだ画像の型と配列を確認してください。
#### 画像はopencv1フォルダにあります。画像の読込にはio.imreadかcv2.imreadを使ってください。
#### 画像の表示はplt.imshowが便利です。
![](./opencv1/water_coins.jpg)

```python
#答え
img = io.imread("./opencv1/water_coins.jpg")
plt.imshow(img)
print(type(img))
print(img.shape)
```

### 問題２：読み込んだ画像をグレースケールにして表示してください。さらに配列の形を確認してください。
#### cv2.cvtColorを使います
![](./opencv1/gray.jpg)

```python
#答え
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,"gray")
print(gray.shape)

cv2.imwrite("./opencv1/gray.jpg",gray)
```

### 問題３：読み込んだ画像を２値化して、コインと背景を白黒に分けて表示してください。
#### cv2.thresholdを使います。
![](./opencv1/binary.jpg)

```python
#答え
ret, binary_img = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
plt.imshow(binary_img,"gray")

cv2.imwrite("./opencv1/binary.jpg",binary_img)
```

### 問題４：問題３で２値化した画像の白黒を反転させて、膨張処理、収縮処理を行ってください。
#### cv2.bitwise_not,cv2.erode,cv2.dilateを使います。
![](./opencv1/bitwise.jpg)
![](./opencv1/erode.jpg)
![](./opencv1/dilate.jpg)

```python
#答え
bitwise_img = cv2.bitwise_not(binary_img)
plt.figure()
plt.imshow(bitwise_img,"gray")
cv2.imwrite("./opencv1/bitwise.jpg",bitwise_img)

kernel = np.ones((8,8),np.uint8)
erode_img = cv2.erode(bitwise_img,kernel)
plt.figure()
plt.imshow(erode_img,"gray")
cv2.imwrite("./opencv1/erode.jpg",erode_img)

dilate_img = cv2.dilate(bitwise_img,kernel)
plt.figure()
plt.imshow(dilate_img,"gray")
cv2.imwrite("./opencv1/dilate.jpg",dilate_img)
```

### 問題５：グレースケール画像を使って、コインの輪郭（エッジ）を検出した画像を表示してください。
#### cv2.Cannyを使います。
![](./opencv1/canny.jpg)

```python
canny_img = cv2.Canny(gray, 120, 150)
plt.imshow(canny_img,"gray")

cv2.imwrite("./opencv1/canny.jpg",canny_img)
```

### 問題６：画像を元画像をx=0:100,y=0:150で切り出して、表示してください。
#### numpy配列の基本操作を使います。
![](./opencv1/cut.jpg)

```python
cut_img = img[0:150,0:100]
plt.imshow(cut_img)

cut_img = cv2.cvtColor(cut_img,cv2.COLOR_RGB2BGR)
cv2.imwrite("./opencv1/cut.jpg",cut_img)
```

### 問題７：画像を150×150にリサイズしてください。
#### cv2.resizeを使います。
![](./opencv1/resize.jpg)

```python
#答え
resize_img = cv2.resize(img,(150,150))
plt.imshow(resize_img)

resize_img = cv2.cvtColor(resize_img,cv2.COLOR_RGB2BGR)
cv2.imwrite("./opencv1/resize.jpg",resize_img)
```

### 問題８：適当な画像を保存してください。
#### cv2.imwriteを使います。

```python
#答え
cv2.imwrite("./opencv1/resize.jpg",resize_img)
```
